```python
import numpy as np
import matplotlib.pyplot as plt
import csv

def filter_outliers(data, m=2):
  ''' remove outliers '''
  std_speed = np.std(data[1,:,0])
  mean_speed = np.mean(data[1,:,0])
  data_no_outliers = data.copy()

  # for i, row in enumerate(data):
    # if row[0] > np.abs(median_speed + m * std_speed) or row[0] > 500:
    #   data_no_outliers = np.delete(data, i, axis=0)
    
  data_no_outliers = data[data[1,:,0] < np.abs(mean_speed + m * std_speed)]

  return data_no_outliers

baseline_data = np.zeros((5, 1000, 2))
a_data = np.zeros((5, 1000, 2))
c_data = np.zeros((5, 1000, 2))
s_data = np.zeros((5, 1000, 2))
sc_data = np.zeros((5, 1000, 2))
as_data = np.zeros((5, 1000, 2))
ac_data = np.zeros((5, 1000, 2))
sca_data = np.zeros((5, 1000, 2))

# Import data
for i in range(1,6):
  CSVData = open(f"data/baseline_{i}.csv")
  baseline_data[i-1] = np.loadtxt(CSVData, delimiter=";")[0:1000,:]
  # baseline_data = filter_outliers(baseline_data)

  CSVData = open(f"data/data-a_{i}.csv")
  a_data[i-1] = np.loadtxt(CSVData, delimiter=";")[0:1000,:]
  # a_data = filter_outliers(a_data)

  CSVData = open(f"data/data-c_{i}.csv")
  c_data[i-1] = np.loadtxt(CSVData, delimiter=";")[0:1000,:]
  # c_data = filter_outliers(c_data)

  CSVData = open(f"data/data-s_{i}.csv")
  s_data[i-1] = np.loadtxt(CSVData, delimiter=";")[0:1000,:]
  # s_data = filter_outliers(s_data)

  CSVData = open(f"data/data-sc_{i}.csv")
  sc_data[i-1] = np.loadtxt(CSVData, delimiter=";")[0:1000,:]
  # sc_data = filter_outliers(sc_data)

  CSVData = open(f"data/data-as_{i}.csv")
  as_data[i-1] = np.loadtxt(CSVData, delimiter=";")[0:1000,:]
  # as_data = filter_outliers(as_data)

  CSVData = open(f"data/data-ac_{i}.csv")
  ac_data[i-1] = np.loadtxt(CSVData, delimiter=";")[0:1000,:]
  # ac_data = filter_outliers(ac_data)

  CSVData = open(f"data/data-sca_{i}.csv")
  sca_data[i-1] = np.loadtxt(CSVData, delimiter=";")[0:1000,:]
  # sca_data = filter_outliers(sca_data)

CSVData = open("data/big_boid.csv")
big_boid = np.loadtxt(CSVData, delimiter=";")[0:1000,:]

CSVData = open("data/100_boids.csv")
x100_boids = np.loadtxt(CSVData, delimiter=";")[0:1000,:]

CSVData = open("data/200_boids.csv")
x200_boids = np.loadtxt(CSVData, delimiter=";")[0:1000,:]

CSVData = open("data/400_boids.csv")
x400_boids = np.loadtxt(CSVData, delimiter=";")[0:1000,:]

CSVData = open("data/500_boids.csv")
x500_boids = np.loadtxt(CSVData, delimiter=";")[0:1000,:]
```
